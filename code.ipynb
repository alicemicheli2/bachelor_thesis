```python
# 3. score function 

def score(path):   
    data = pd.read_csv(path, index_col=0) #read the csv file
    ordered_data = data.sort_values('scores', ascending=False)\
    .drop(data[data['scores'] < 0.001].index).reset_index()   # Orders by score and drops nets with score < 0.001
   
    net = 0 #best network is the first of the new ordered data 
    x= (ordered_data['scores'][0]) #GET THE SCORE 
    return x
```

```python
#4. parameterizing_10_p_dictionary function 

def parameterizing_10p_dictionary (path):    
    import glob
    import pandas as pd
    import numpy as np

    networks=[]
    score=[] 
    ten_perc=[]
    
    data = pd.read_csv(path, index_col=0) #read csv file 
    ordered_data = data.sort_values('scores', ascending=False).reset_index() # Orders by score
 
    for i in range(20): #best 10% of 200 networks 
        net = i  
        networks.append(i)#creating the netowrks list
        sco= (ordered_data['scores'][i])
        score.append(sco) #creating the score list 
        
        IN_list = np.array(eval(ordered_data['INs'][net]))[:, 0]
        input_syn = np.array(eval(ordered_data['input_syn'][net]))[:, :2]
        output_syn = np.array(eval(ordered_data['output_syn'][net]))[:, :2]

        final_list=np.zeros(54) #create an array of 54 elements with all initial value equal to zero 

        for inhibitor in IN_list: #access each inhibitory neuron 
            final_neurons=[]
            inhibitor_output= []
            inhibitor_input=[]

        #OUTPUT ---------------------------------------------------------------------------#
        #access the output value with that inhibitor number 

            for i in output_syn:    
                if inhibitor==i[0]:
                    inhibitor_output.append(i)
            inhibitor_output=np.float64(inhibitor_output)#need to change to float 64 becuase of an error that might occur 

        # change the value of inhibitory neuron to value 100  
            for k in range(len(inhibitor_output)):
                inhibitor_output[k][0]=100
        #-----------------------------------------------------------------------------------#
        
        
        
        #INPUT -----------------------------------------------------------------------------#
        #access the input value with that inhibitory neuron  number      
            for j in input_syn:
                if inhibitor == j[1]:
                    inhibitor_input.append(j)
            inhibitor_input=np.float64(inhibitor_input)

        # change the value of inhibitory neuron to value 100 
            for n in range(len(inhibitor_input)):
                inhibitor_input[n][1]=100 
        #-----------------------------------------------------------------------------------# 
        
        
        
        # create one single list with edges of directed graph  
            for x in inhibitor_input:
                inhibitor_output = np.append(inhibitor_output,[x], axis=0) 
                final_neurons=inhibitor_output
                final_neurons_changed=transform_matrix(final_neurons)
                
            #add +1 to the entry of the matrix according to presence of such motif 
            final_list[index_of_small_graph(final_neurons_changed)] = final_list[index_of_small_graph(final_neurons_changed)]+1

            ten_perc.append(final_list) #ten_perc is a 20x54 matrix for each generation 
    return ten_perc, networks, score #return the matrix, the list of netowrk, and the list of scores 
```

```python
# 5. find_PC function

from sklearn.decomposition import PCA 

def find_PC(path):
    a=parameterizing_10p_dictionary(path)[0]#matrix 20x54
    pca=PCA(n_components=0.95)#perform PCA
    pca.fit(a)
    return np.shape(pca.transform(a))[1] # matrix 20xN, return N
```

```python
# 6. parameterizing_10p_dictionary_worse function  

def parameterizing_10p_dictionary_worse (path):    
    import glob
    import pandas as pd
    import numpy as np

    networks=[]
    score=[]
    ten_perc=[]
     
    data = pd.read_csv(path, index_col=0) #read csv file 
    ordered_data = data.sort_values('scores', ascending=False)\
    [::-1].reset_index()#sort the order of the data
    
    for i in range(20): #best 10% of 200 networks 
        net = i 
        networks.append(i) #get the list of networks 
        sco= (ordered_data['scores'][i])
        score.append(sco) #get the list of the scores 
        
        try: 
            IN_list = np.array(eval(ordered_data['INs'][net]))[:, 0]
            input_syn = np.array(eval(ordered_data['input_syn'][net]))[:, :2]
            output_syn = np.array(eval(ordered_data['output_syn'][net]))[:, :2]
            
        except:
            IN_list = np.array([])
            input_syn =np.array([])
            output_syn= np.array([])
            
        #create an array of 54 elements with all initial value equal to zero 
        final_list=np.zeros(54)

        for inhibitor in IN_list: #access each inhibitory neuron 
            final_neurons=[]
            inhibitor_output= []
            inhibitor_input=[]

        #OUTPUT ---------------------------------------------------------------------------#
        #access the output value with that inhibitory neuron number 

            for i in output_syn:    
                if inhibitor==i[0]:
                    inhibitor_output.append(i)
            inhibitor_output=np.float64(inhibitor_output)#need to change to float 64 becuase of an error that might occur 

        # use a loop to change the index of the inhibitor to value 100  
            for k in range(len(inhibitor_output)):
                inhibitor_output[k][0]=100
        #-----------------------------------------------------------------------------------#
        
        
        
        
        
        #INPUT -----------------------------------------------------------------------------#
        #access the input value with that inhibitory neuron number      
            for j in input_syn:
                if inhibitor == j[1]:
                    inhibitor_input.append(j)
            inhibitor_input=np.float64(inhibitor_input)

        #  change the value of the inhibitory neuron to value 100 
            for n in range(len(inhibitor_input)):
                inhibitor_input[n][1]=100 
        #-----------------------------------------------------------------------------------#
        
        
        
        # create one single list with edges of directed graph  
            for x in inhibitor_input:
                inhibitor_output = np.append(inhibitor_output,[x], axis=0) 
                final_neurons=inhibitor_output
                final_neurons_changed=transform_matrix(final_neurons)
            #add +1 to the entry of the matrix according to presence of such motif 
            final_list[index_of_small_graph(final_neurons_changed)] = final_list[index_of_small_graph(final_neurons_changed)]+1
        ten_perc.append(final_list) #ten_perc is a 20x54 matrix for each generation 
    return ten_perc, networks, score  #return the matrix, the list of netowrk, and the list of scores 
```

```python
# 7. find_PC_worse function  

from sklearn.decomposition import PCA 

def find_PC_worse(path):
    a=parameterizing_10p_dictionary_worse (path)[0]#matrix 20x54
    pca=PCA(n_components=0.95) #perform PCA 
    pca.fit(a)
    return np.shape(pca.transform(a))[1] # matrix 20xN, return N
```

```python
import glob
import pandas as pd
import numpy as np
generation=[]
simulation=[]
n_inhibitors=[]
networks=[]
topology_two=[]
topology_three=[]
score=[]

list_simulation=['1','2','3','4','5','6','7','8']

for element_j in list_simulation: #for every simulation
    hash = element_j #simulation 
    run=[]
    net=[]
    sco=[]
    for i in range(501):#for all generations 
        gen = i
        for f in glob.glob('nets/{}*/{}.csv'.format(hash, gen)):  # pathnames
            path = f
            path=str(path)
            run.append(parameterizing_10p_dictionary (path)[0])
            net.append(parameterizing_10p_dictionary (path)[1])
            sco.append(parameterizing_10p_dictionary (path)[2])
       
    for k in run:
        for m in k:
            current_run=np.array(m)
            current_inhibitor=np.sum(current_run)
            n_inhibitors.append(current_inhibitor)
            
            of_two= current_run[0:27]# it is the first 27 entries (so from entry 0 to entry 26)
            of_three=current_run[27:] #it is the last 27 entries ( so from entry 27 to entry 53)
            
            d=np.sum(of_two>0) #to check if there are non-zero entries  in the first 27 entries
            e=np.sum(of_three>0) #to check if there are non-zero entries in the last 27 entries 
            
            topology_two.append(d)  
            topology_three.append(e)
            
        for n in range(20):
            simulation.append(element_j) #create simulation list 
            
    for element in net: 
        for l in element: 
            networks.append(l) #create networks list 
            
    for element in sco: 
        for l in element: 
            score.append(l)#create score list 
        
# for PC ------------------------------------------------------------------------------------#
PC_star=[]
gen_PC=[]
sim_PC=[]

for element_j in list_simulation: 
    hash = element_j #simulation

    for i in range (501): #for every generation 
        gen = i
        for f in glob.glob('nets/{}*/{}.csv'.format(hash, gen)):  # Lists all pathnames
            path = f
            path=str(path)
            
            PC_star.append(find_PC(path)) #append N 
            gen_PC.append(i) #create generation list for PC
            sim_PC.append(element_j) #create simulation list for PC 
            
# generation -------------------------------------------------------------------------------#
for element_j in (list_simulation):
    for i in range (501):
        for j in range(20):
            generation.append(i)#create generation list 

# the final lists that we have are generation, simulation, n_inhibitory neurons, 
# networks, simple motifs, complex motifs, score, PCs
print('complete')
```
